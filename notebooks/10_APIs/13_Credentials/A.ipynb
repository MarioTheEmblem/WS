# API Credentials

## Design API call

```python
import requests
import pandas as pd
```

```python
#Con el ejemplo proporcionado, la parametrización de la URL sería...

URL_base = 'https://www.alphavantage.co' #Se considera solo lo previo al primer slash y sin incluir a este.

End_point = '/query' #Se considera solo lo previo al signo de interrogación.

URL = URL_base + End_point
```

```python
#Son todos elementos dentro de una URL que estan separados por un ampersand ("&"), se definen como diccionarios.
params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': 'IBM',
    'outputsize': 'full',
    'apikey': 'demo'
}
```

```python
#Hasta ahora ha sido el mismo proceso, aquí cambiaremos el valor de un parámetro, buscando obtener la información
#sobre las acciones de Banco Santander (SAN), o para cualquier otra acción que cotize en bolsa de valores.
params['symbol'] = 'SAN'
```

```python
res = requests.get(URL,params)
res #Aun que nos muestre un Código 200 en realidad no se podra visualizar la respuesta, dado que la apikey es demo.
```

```python
res.json() #Aqui precisamente nos señala que nuestras opciones con un token demo son limitadas, nos registrarémos para ampliar dichas opciones.
```
